### 3. Ollama Server im Hintergrund starten

```python
import subprocess
import time

# Eventuell laufende Instanzen killen
subprocess.run(["pkill", "-f", "ollama"], capture_output=True)
time.sleep(1)

# Server starten
subprocess.Popen(["/usr/local/bin/ollama", "serve"], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
time.sleep(10)
print("Ollama Server läuft im Hintergrund.")
```

### 6. Rat der Titanen (Runde 2) ausführen

Das Skript läuft sequentiell durch alle Titanen. Dauer ca. 60-90 Minuten.

```python
import os
import subprocess

os.environ["OLLAMA_LOCAL_HOST"] = "http://127.0.0.1:11434"
os.environ["OMEGA_COUNCIL_MODEL"] = "qwen2.5:14b"
os.environ["OMEGA_COUNCIL_NUM_CTX"] = "65536"

print("Starte Science Council Runde 2...")
# Wir nutzen --force um sicherzugehen, dass alles neu generiert wird
subprocess.run(["python3", "src/scripts/run_omega_science_council_r2.py", "--force"])
```
